```python
import pandas as pd
import csv

# opening the CSV file for input dataset and preprocessing by removing unwanted columns
with open('news_summary.csv', mode ='r')as file:     
  # reading the CSV file 
    csvFile = csv.reader(file)
    df = pd.DataFrame(csvFile)
    
df.columns = ["author","date","headlines","read_more","text_summary","text"]
df = df.drop(0)  

df = df.drop(['author',"date","headlines","read_more"], axis = 1) 

df.head()
len(df)
```

```python
from summarizer import  Summarizer
```

```python
bert_model = Summarizer()
```

```python
# Text preprocessing
import re
preprocessed_text = [] 
preprocessed_references = []
for i in range(1,len(df)):
    preprocessed_text.append(re.sub('[^A-Za-z0-9.]+', ' ',df["text"][i]))
    preprocessed_references.append(re.sub('[^A-Za-z0-9.]+', ' ',df["text_summary"][i]))
```

```python
# Generating the summaries for the pre processed text
references = []
hypothesis = []
for i in range(1,len(df)):
    references.append(bert_model(preprocessed_text[i]) 
    hypothesis.append(preprocessed_references[i])
    
```
